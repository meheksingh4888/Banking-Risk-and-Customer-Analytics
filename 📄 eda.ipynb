```python
total_transactions = len(transactions)

failed_transactions = (
    transactions["status"] == "Failed"
).sum()

failed_percentage = (
    failed_transactions / total_transactions
) * 100

print("Failed Transactions:", failed_transactions)
print("Failure Rate:", round(failed_percentage,2), "%")
```

```python
plt.figure(figsize=(6,4))

transactions["status"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Transaction Success vs Failure")

plt.ylabel("")

plt.show()
```

```python
print("="*70)
print("BANKING RISK CUSTOMER ANALYTICS - BUSINESS INSIGHTS")
print("="*70)


# Customer Insights

print("\nCUSTOMER INSIGHTS")

print(
    "Total Customers:",
    len(customers)
)

print(
    "Highest Customer State:",
    customers["state"].value_counts().idxmax()
)

print(
    "Average Credit Score:",
    round(customers["credit_score"].mean(),2)
)


# Transaction Insights

print("\nTRANSACTION INSIGHTS")

print(
    "Total Transactions:",
    len(transactions)
)

print(
    "Total Transaction Volume: ₹",
    round(transactions["amount"].sum(),2)
)

print(
    "Most Used Channel:",
    transactions["channel"].value_counts().idxmax()
)


# Loan Insights

print("\nLOAN INSIGHTS")

print(
    "Total Loans:",
    len(loans)
)

print(
    "Total Loan Portfolio: ₹",
    round(loans["loan_amount"].sum(),2)
)


# Fraud Insights

print("\nFRAUD INSIGHTS")

failed = (
    transactions["status"]=="Failed"
).sum()

print(
    "Failed Transactions:",
    failed
)

print(
    "Fraud Failure Rate:",
    round((failed/len(transactions))*100,2),
    "%"
)


print("\n")
print("="*70)
print("EDA COMPLETED SUCCESSFULLY")
print("="*70)
```
